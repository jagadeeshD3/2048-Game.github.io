```python
df = df[["coaster_name","Manufacturer","year_introduced","latitude","longitude","Type_Main","opening_date_clean","speed_mph",
    "height_value","height_unit","height_ft","Inversions_clean","Gforce_clean","Location","Status"]].copy()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df["opening_date_clean"] = pd.to_datetime(df["opening_date_clean"])
```

```python
df["Gforce_clean"] = pd.to_numeric(df["Gforce_clean"])
```

```python
df.columns
```

```python
#renaming columns
df = df.rename(columns={
    "coaster_name":"Coaster_Name",
    "year_introduced":"Year_Introduced",
    "latitued":"Latitude",
    "longitude":"Longitude",
    "opening_date_clean":"Opening_Date",
    "speed_mph":"Speed_Mph",
    "height_value":"Height_Value",
    "height_unit":"Height_Unit",
    "height_ft":"Height_Ft",
    "Inversions_clean":"Inversions",
    "Gforce_clean":"Gforce",
}).copy()
```

```python
df = df.loc[~df.duplicated(["Coaster_Name","Location","Opening_Date"])]\
.reset_index(drop=True)\
.copy()
```
